```python
# Plot the transmission Range against Efficiency for Capacity = 1  & Sim. Duration = 0.1 min.
for i in data:
    plt.plot(i.loc[i["Trans Capacity"] == 1, "Trans Range"], 
             i.loc[i["Trans Capacity"] == 1, "Trans Efficiency"])
    
plt.xlabel("Transmission Range (Feet)")
plt.ylabel("Transmission Efficiency (%)")

plt.title("Ten Runs For Transmission Capacity = 1")
plt.show()

# Plot the transmission Range against Efficiency for Capacity = 2
for i in data:
    plt.plot(i.loc[i["Trans Capacity"] == 2, "Trans Range"], 
             i.loc[i["Trans Capacity"] == 2, "Trans Efficiency"])
    
plt.xlabel("Transmission Range (Feet)")
plt.ylabel("Transmission Efficiency (%)")

plt.title("Ten Runs For Transmission Capacity = 2")
plt.show()

# Plot the transmission Range against Efficiency for Capacity = 3
for i in data:
    plt.plot(i.loc[i["Trans Capacity"] == 3, "Trans Range"], 
             i.loc[i["Trans Capacity"] == 3, "Trans Efficiency"])
    
plt.xlabel("Transmission Range (Feet)")
plt.ylabel("Transmission Efficiency (%)")

plt.title("Ten Runs For Transmission Capacity = 3")
plt.show()

# Plot the transmission Range against Efficiency for Capacity = 4
for i in data:
    plt.plot(i.loc[i["Trans Capacity"] == 4, "Trans Range"], 
             i.loc[i["Trans Capacity"] == 4, "Trans Efficiency"])
    
plt.xlabel("Transmission Range (Feet)")
plt.ylabel("Transmission Efficiency (%)")

plt.title("Ten Runs For Transmission Capacity = 4")
plt.show()

# Plot the transmission Range against Efficiency for Capacity = 5
for i in data:
    plt.plot(i.loc[i["Trans Capacity"] == 5, "Trans Range"], 
             i.loc[i["Trans Capacity"] == 5, "Trans Efficiency"])
    
plt.xlabel("Transmission Range (Feet)")
plt.ylabel("Transmission Efficiency (%)")

plt.title("Ten Runs For Transmission Capacity = 5")
plt.show()
```
